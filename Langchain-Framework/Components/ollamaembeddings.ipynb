```python
from langchain_community.embeddings import OllamaEmbeddings
ollama_embeddings = (
    OllamaEmbeddings(model="gemma:2b", base_url="http://localhost:11434")
)
ollama_embeddings
```

```python
r1 = ollama_embeddings.embed_documents(
    [
"LangChain is a framework for developing applications powered by language models.",
"LangChain provides a standard interface for connecting to various language models and tools."
    ]
)
r1
```
